```python
import numpy as np
import pandas as pd 
import matplotlib.pyplot as plt 
import seaborn as sns
```

```python
df=pd.read_csv('accident.csv')
df.head()
```

```python
graph = sns.countplot(x='Survived',data=df,hue='Survived')
plt.title("Survived Counts")
plt.xlabel("Survived")
plt.ylabel("Counts")

for container in graph.containers:
    graph.bar_label(container)

plt.show()
```

```python
graph = sns.countplot(x='Helmet_Used',data=df,hue='Helmet_Used')
plt.title("Helmet_Used count")
plt.xlabel("Helmet_Used")
plt.ylabel("Count")

for container in graph.containers:
    graph.bar_label(container)

plt.show()
```

```python
graph = sns.countplot(x='Seatbelt_Used',data=df,hue='Seatbelt_Used')
plt.title("Seatbelt_Used count")
plt.xlabel("Seatbelt_Used")
plt.ylabel("Count")

for container in graph.containers:
    graph.bar_label(container)

plt.show()
```

```python
graph = sns.countplot(x='Helmet_Used',hue = 'Seatbelt_Used', data=df)
plt.title('Helmet_Used vs Seatbelt_Used')

for container in graph.containers:
    graph.bar_label(container)

plt.show()
```

```python
print('Oldest passenger was of: ',df['Speed_of_Impact'].max(),'Speed_of_Impact')
print('Youngest passenger was of: ',df['Speed_of_Impact'].min(),'Speed_of_Impact')
print('Average passenger was of: ',df['Speed_of_Impact'].mean(),'Speed_of_Impact')

plt.figure(figsize=(18,8))

sns.histplot(df['Speed_of_Impact'],bins=np.arange(df['Speed_of_Impact'].min(),df['Speed_of_Impact'].max(),5),color='black',edgecolor='red',kde=True)
plt.title('Speed_of_Impact at Accidents')
plt.xlabel('Speed_of_Impact')
plt.ylabel('Counts')
plt.show()
```

```python
graph = sns.countplot(x='Gender',hue = 'Survived', data=df)
plt.title('Gender vs Survived')

for container in graph.containers:
    graph.bar_label(container)

plt.show()
```
