```python
import pandas as pd
df=pd.read_csv("Samsung.csv")
df.head()
```

```python
data=df[['Date','Open','High','Low','Close','Adj Close','Volume']]
```

```python
data.info()
```

```python
data['Date']=pd.to_datetime(data['Date'])

#set 'data' column as the index
data.set_index('Date',inplace=True)

data.head()
```

```python
import matplotlib.pyplot as plt

data['Open'].plot(figsize=(10,6),title='Samsung Stock Prices')
plt.ylabel('Open')
```

```python
import matplotlib.pyplot as plt

data['High'].plot(figsize=(10,6),title='Samsung Stock (High) Prices')
plt.ylabel('High')
```
