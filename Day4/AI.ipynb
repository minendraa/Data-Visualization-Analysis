```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
```

```python
df=pd.read_csv('ai_job_dataset.csv')
```

```python
df.head(10)
```

```python
df.isnull().sum()
```

```python
graph=sns.countplot(x='salary_currency',data=df,hue='salary_currency')
plt.title('---Currency variance Counts---')
plt.xlabel('Currency')
plt.ylabel('Count')

for container in graph.containers:
    graph.bar_label(container)

plt.show()
```

```python
graph=sns.countplot(x='company_size',data=df,hue='company_size')
plt.title('---Company size variance Counts---')
plt.xlabel('Company Size')
plt.ylabel('Count')

for container in graph.containers:
    graph.bar_label(container)

plt.show()
```

```python
graph=sns.countplot(x='education_required',data=df,hue='education_required')
plt.title('---Education Required variance Counts---')
plt.xlabel('Education Type')
plt.ylabel('Count')

for container in graph.containers:
    graph.bar_label(container)

plt.show()
```

```python
sns.scatterplot(data=df,x='years_experience',y='salary_usd',color="red")

plt.title('Experience vs Salary')
plt.xlabel('experience_years')
plt.ylabel('salary_usd')
plt.show()
```

```python
plt.figure(figsize=(18,8))
sns.histplot(df['years_experience'],bins=np.arange(df['years_experience'].min(),df['years_experience'].max(),5),color='steelblue',edgecolor='black',kde=True)
plt.title('Experience Frequency')
plt.xlabel('years_experience')
plt.ylabel('Frequency')
plt.show()
```

```python
plt.figure(figsize=(18,8))
sns.histplot(df['salary_usd'],bins=np.arange(df['salary_usd'].min(),df['salary_usd'].max(),5),color='steelblue',edgecolor='black',kde=True)
plt.title('Salary Frequency')
plt.xlabel('Salary')
plt.ylabel('Frequency')
plt.show()
```
