```python
import numpy as np
import pandas as pd 
import matplotlib.pyplot as plt 
import seaborn as sns 
```

```python
df=pd.read_csv('drivers_updated.csv')
df.tail()
```

```python
df['Car'].value_counts()
```

```python
df['Pos'].value_counts()
```

```python
sns.scatterplot(data=df,x='year',y='PTS',color='green')
plt.yticks(range(0, 601, 25))

plt.show()
```

```python
plt.figure(figsize=(18,8))
sns.histplot(df['PTS'],bins=np.arange(df['PTS'].min(),df['PTS'].max(),5),color='steelblue',edgecolor='black',kde=True)
plt.title('Experience Frequency')
plt.xlabel('Points')
plt.ylabel('Frequency')
plt.show()
```
